```python
### some useful class definitions
class AntNode:
    def __init__(self, name):
        self.name = name
        self.edges = []
        self.edgeWeights = {}
        self.numAnts = 0
        self.q_decay = 0.0
        self.isStart = False
        self.isGoal = False
        self.visited = False
        
    def __hash__(self):
        return hash(self.name)

    def __eq__(self, other):
        return hash(self)==hash(other)

    def __neq__(self, other):
        return self==other
    
    def addEdge(self, node, weight):
        self.edges.append(node)
        self.edgeWeights[node] = weight
    
    def decayEdges(self):
        for e in self.edgeWeights.keys():
                self.edgeWeights[e] *= (1-self.q_decay)

class Ant:
    def __init__(self):
        self.currentNode = None
        self.pastNode = None
        self.q_explore = 0.0
        self.q_add = 0.0
        self.backtracking = False
    
    def move(self):
        
        if(self.backtracking):
            neighbors = []
            for n in self.currentNode.edges:
                if(not n == self.pastNode):
                    neighbors.append(n)
            if(len(neighbors) >= 2):
                backtracking = False
            elif(len(neighbors) == 0):
                self.backtracking = True
                temp = self.currentNode
                self.currentNode = self.pastNode
                self.pastNode = temp
                self.pastNode.numAnts -= 1
                self.currentNode.numAnts += 1
                return
            else:
                self.pastNode = self.currentNode
                self.currentNode = neighbors[-1]
                self.pastNode.numAnts -= 1
                self.currentNode.numAnts += 1
                return
            
        neighbors = []
        for n in self.currentNode.edges:
            if(not n == self.pastNode):
                neighbors.append(n)
        random.shuffle(neighbors)
        
        ## if at a dead end
        if(len(neighbors) == 0 and not (self.currentNode.isGoal or self.currentNode.isStart)):
            self.backtracking = True
            temp = self.currentNode
            self.currentNode = self.pastNode
            self.pastNode = temp
            return
        else:
            if(len(neighbors) == 0):
                neighbors.append(self.pastNode)
            neighbors.sort(key=lambda x: self.currentNode.edgeWeights[x], reverse=True)
            roll = random.random()
            moved = False
            index=0
            for n in neighbors:
                if(roll < 1-self.q_explore):
                    ## dont explore
                    self.pastNode = self.currentNode
                    self.currentNode = n
                    self.pastNode.numAnts -= 1
                    self.currentNode.numAnts += 1
                    
                    if(self.pastNode.edgeWeights[n] - self.pastNode.edgeWeights[neighbors[0]] > tol):
                        ## move back from explore
                        temp = self.currentNode
                        self.currentNode = self.pastNode
                        self.pastNode = temp
                        self.pastNode.numAnts -= 1
                        self.currentNode.numAnts += 1
                        self.pastNode.edgeWeights[self.currentNode] += self.q_add
                        self.currentNode.edgeWeights[self.pastNode] += self.q_add
                    else:
                         moved = True
                    
                    break
                index +=1
            if(not moved):
                self.pastNode = self.currentNode
                self.currentNode = neighbors[-1]
                self.pastNode.numAnts -= 1
                self.currentNode.numAnts += 1
                ## move back from explore
                temp = self.currentNode
                self.currentNode = self.pastNode
                self.pastNode = temp
                self.pastNode.numAnts -= 1
                self.currentNode.numAnts += 1
                self.pastNode.edgeWeights[self.currentNode] += self.q_add
                self.currentNode.edgeWeights[self.pastNode] += self.q_add
            else:
                self.pastNode.edgeWeights[self.currentNode] += self.q_add
                self.currentNode.edgeWeights[self.pastNode] += self.q_add
```

```python
### generates a sample of a network using bfs
def bfsSampler(edgeList, sampleSize):
    edgeTable = {}
    nodeSet = set()
    nodeList = []
    for e in edgeList:
        if(e[0] not in edgeTable.keys()):
            edgeTable[e[0]] = []
        edgeTable[e[0]].append(e[1])
        nodeSet.add(e[0])
        nodeSet.add(e[1])
    nodeList = list(nodeSet)
        
    finished = False
    while not finished:
        ### build the sample graph
        visitedNodes = set()
        sampleEdges = []
        sampleNodes = set()
        currentNode = random.choice(nodeList)
        sampleNodes.add(currentNode)
        visitedNodes.add(currentNode)
        
        if(currentNode not in edgeTable.keys()):
            continue
        
        for e in edgeTable[currentNode]:
            if(len(sampleNodes) < sampleSize):
                if((currentNode,e) not in sampleEdges):
                    ### visit node
                    sampleEdges.append((currentNode, e))
                    sampleNodes.add(e)
                    bfs_samp_visit(e, 
                                   sampleEdges, 
                                   sampleNodes,
                                   visitedNodes,
                                  edgeTable,
                                  sampleSize)
                if(len(sampleNodes) >= sampleSize):
                    break
            else:
                break
        if(len(sampleNodes) >= sampleSize):
            finished=True
    
    return sampleEdges
                    
def bfs_samp_visit(n, 
                   sampleEdges, 
                   sampleNodes, 
                   visitedNodes, 
                   edgeTable,
                   sampleSize):
    visitedNodes.add(n)
    
    if(n not in edgeTable.keys()):
        return
    
    for e in edgeTable[n]:
        if(len(sampleNodes) < sampleSize):
            if((n,e) not in sampleEdges):
                sampleEdges.append((n,e))
                sampleNodes.add(e)
                bfs_samp_visit(e, 
                               sampleEdges, 
                               sampleNodes,
                               visitedNodes,
                              edgeTable,
                              sampleSize)
                if(len(sampleNodes) >= sampleSize):
                    break
        else:
            break

### returns all paths of an undirected graph
def bfsPathFinder(edgeList, start, goal):
    edgeTable = {}
    nodeSet = set()
    nodeList = []
    for e in edgeList:
        if(e[0] not in edgeTable.keys()):
            edgeTable[e[0]] = []
        edgeTable[e[0]].append(e[1])
        nodeSet.add(e[0])
        nodeSet.add(e[1])
    nodeList = list(nodeSet)
    
    ### find paths
    paths = []
    visitedNodes = set()
    currentNode = start
    if(currentNode in edgeTable.keys()):
        visitedNodes.add(currentNode)

        for e in edgeTable[currentNode]:
            if(e not in visitedNodes and e!=currentNode):
                ### visit node
                newPaths = bfsPFVisit(e, 
                                      visitedNodes, 
                                      edgeTable, 
                                      [currentNode],
                                         goal)
                paths.extend(newPaths)
    
    return paths

def bfsPFVisit(node, visitedNodes, edgeTable, p, goal):
    path = []
    for n in p:
        path.append(n)
    path.append(node)
    if(path[-1] == goal):
        return [path]
    visitedNodes.add(node)
    paths = []
    if(node in edgeTable.keys()):
        for n in edgeTable[node]:
            if(not n in visitedNodes and n!=node):
                paths.extend(bfsPFVisit(n, visitedNodes, edgeTable, path, goal))
    return paths
            
### automatic scenario generation for given edge list
def generateScenario(edgeList):
    finished = False
    originalPath = []
    start = 0
    goal = 0
    notTried = []
    nodeSet = set()
    for e in edgeList:
        nodeSet.add(e[0])
        nodeSet.add(e[1])
    for n in nodeSet:
        for m in nodeSet:
            if(n != m):
                notTried.append((n,m))
    random.shuffle(notTried)
    i = 0
    while not finished:

        if(i >= len(notTried)):
            print("Tried everything")
            return
        
        start = notTried[i][0]
        goal = notTried[i][1]
        initPaths = bfsPathFinder(edgeList,start,goal)
        if(len(initPaths) >= 2):
            print("Candidate:")
            print(initPaths)
            ### remove a random edge
            modEdgeList = []
            originalPath = initPaths[0]
            edgeIndex = random.randint(0,len(initPaths[0])-3)
            edgeToRemove = (initPaths[0][edgeIndex], 
                            initPaths[0][edgeIndex+1])
            for e in edgeList:
                if(e != edgeToRemove):
                    modEdgeList.append(e)
            newPaths = bfsPathFinder(modEdgeList,start,goal)
            print("removed " + str(edgeToRemove) + ", " + str(len(newPaths)))
            if(len(newPaths) > 0):
                pathLengths = []
                for p in newPaths:
                    pathLengths.append(len(p))
                minPathLength = np.min(pathLengths)
                finished = True
        i += 1
    return [start, goal, modEdgeList, originalPath, minPathLength, newPaths[0], edgeToRemove]
```

```python
def bfs_all(startNode):
    
    startNode.visited=True
    
    paths = []
    
    for n in startNode.edges:
        if(startNode.edgeWeights[n] > tol and not n.visited):
            newPaths = bfsvisit(n, [startNode])
            paths.extend(newPaths)
    
    return paths

def bfsvisit(node, p):
    path = []
    for n in p:
        path.append(n)
    path.append(node)
    if(path[-1].isGoal):
        return[path]
    node.visited = True
    paths = []
    for n in node.edges:
        if(node.edgeWeights[n] > tol and not n.visited):
            paths.extend(bfsvisit(n, path))
    return paths


def calcSuccess(node):
    currentNode = node
    currentNode.visited = True
    if(currentNode.isGoal):
        return 1.0
    else:
        i = 0
        for n in currentNode.edges:
            if(currentNode.edgeWeights[n] > tol and not (n.visited)):
                if(n.isGoal):
                    return 1.0
                else:
                    return calcSuccess(n)
            i+=1
    return 0.0
```

```python
from networkx.drawing.nx_agraph import graphviz_layout

def getlayout(graph, testname):
    edgeSet = set()
    weightDict = {}
    antCountsDict = {}
    for n in graph:
        antCountsDict[n.name] = n.numAnts
        for e in n.edges:
            edgeNodes = [int(n.name), int(e.name)]
            edgeNodes.sort()
            edgeTuple = (edgeNodes[0], edgeNodes[1])
            edgeSet.add(edgeTuple)
            weightDict[edgeTuple] = n.edgeWeights[e]
    g = nx.Graph()
    weightList = []
    nodeSet = set()
    for e in edgeSet:
        g.add_edge(e[0], e[1])
        nodeSet.add(e[0])
        nodeSet.add(e[1])
        if(max(list(weightDict.values())) > 0.0):
            weightList.append(1.0 + 5.0*(weightDict[e])/(max(list(weightDict.values()))))
        else:
            weightList.append(1.0)
    
    if(testname == "fullGrid" or testname == "fullGrid_final"):
        l = nx.spectral_layout
        return l(g)
    else:
        l = graphviz_layout
        return l(g, prog="neato")


def graphViewer(graph, testname, layout):
    edgeSet = set()
    weightDict = {}
    antCountsDict = {}
    colorsDict = {}
    for n in graph:
        antCountsDict[n.name] = n.numAnts
        if(n.isStart):
            colorsDict[n.name] = "blue"
        elif(n.isGoal):
            colorsDict[n.name] = "green"
        else:
            colorsDict[n.name] = "red"
        for e in n.edges:
            edgeNodes = [int(n.name), int(e.name)]
            edgeNodes.sort()
            edgeTuple = (edgeNodes[0], edgeNodes[1])
            edgeSet.add(edgeTuple)
            weightDict[edgeTuple] = n.edgeWeights[e]
    g = nx.Graph()
    weightList = []
    colorList = []
    nodeSet = set()
    for e in edgeSet:
        g.add_edge(e[0], e[1])
        
    for e in g.edges():
        nodeSet.add(e[0])
        nodeSet.add(e[1])
        if(e not in weightDict.keys()):
            weightList.append(1.0)
            continue
        if(max(list(weightDict.values())) > 0.0):
            weightList.append(1.0 + 10*(weightDict[e])/(max(list(weightDict.values()))))
        else:
            weightList.append(1.0)
    
    antCounts = []
    for n in g.nodes():
        antCounts.append(10.0 + antCountsDict[n])
        colorList.append(colorsDict[n])
        
    plt.figure()
    plt.title(testname)
    nx.draw(g, pos=layout, node_size=antCounts, width=weightList, node_color=colorList, with_labels=False)
    plt.savefig("./plots/" + testname + ".png")
```
